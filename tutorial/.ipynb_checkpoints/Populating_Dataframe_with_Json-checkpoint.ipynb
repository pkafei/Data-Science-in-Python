# Populating Dataframes with JSON

```python
import urllib2
import json
import pandas as pd 
import numpy as np
from IPython.display import Image
import matplotlib.pyplot as plt 
```

The urllib2 module that opens urls and accepts objects. Json.load decodes json objects into Python objects.


```python
# Data source is provided by http://data.cityofnewyork.us/resource/s7yj-m732.json
doe_json = urllib2.urlopen('https://data.cityofchicago.org/api/views/xzkq-xp2w/rows.json?accessType=DOWNLOAD')
doe = json.load(doe_json)
```

Let us check the type of doe, using the doe function

```python
type(doe)
```

```python
## We are going to create a dataframe using json from the city of New York. 
```

```python
department_education = pd.DataFrame(columns=['Position Title', 'Department', 'Employee Annual Salary'])

for x in doe['data']:
    department_education.loc[len(department_education)] = [x['Position Title'], x['Department'], x['Employee Annual Salary']]
```
